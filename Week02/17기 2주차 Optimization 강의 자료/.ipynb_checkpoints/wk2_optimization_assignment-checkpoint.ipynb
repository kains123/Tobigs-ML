## Dot product
## $z = X_i \theta$

```python
def dot_product(X, parameters):
    z = np.dot(X, parameters)
    return z
```

## Logistic Function

## $p(X_i)= \cfrac{1}{1+e^{-X_i \theta}}$

```python
def logistic(X, parameters):
    p = 1 / (1 + np.exp(-dot_product(X, parameters)))
    return p
```

## Object function

Object Function : 목적함수는 Gradient Descent를 통해 최적화 하고자 하는 함수입니다.  
<br>
선형 회귀의 목적함수
## $l(\theta) = \frac{1}{2}\Sigma(y_i - \theta^{T}X_i)^2$  
참고) $\hat{y_i} = \theta^{T}X_i$
  
로지스틱 회귀의 목적함수를 작성해주세요  
(선형 회귀의 목적함수처럼 강의에 나온대로 작성해주세요. 평균을 고려하는 것은 뒤에 코드에서 수행합니다)
## $l(p) = - \Sigma(y_iX_i\theta + log(p))$

```python
def minus_log_cross_entropy_i(X, y, parameters):
    p = logistic(X, parameters)
    loss = -y*np.log(p) - (1-y)*np.log(1-p)
    return loss
```

```python
def mse_i(X, y, parameters):
    
    y_hat = dot_product(X, parameters)
    loss = ((y - y_hat)**2) * 0.5
    return loss
```

```python
def batch_loss(X_set, y_set, parameters, loss_function, n): #n:현재 배치의 데이터 수
    loss = 0
    for i in range(X_set.shape[0]):
        X = X_set.iloc[i,:]
        y = y_set.iloc[i]
        loss += loss_function(X, y, parameters)
    loss = loss/n
    return loss
```

## Gradient
위의 선형회귀의 목적함수 $l(\theta)$와 로지스틱회귀의 목적함수 $l(p)$의 gradient를 작성해주세요  
(위의 목적함수를 참고해서 작성해주세요 = 평균을 고려하는 것은 뒤에 코드에서 수행합니다)

## ${\partial\over{\partial \theta_j}}l(\theta)= -\Sigma(y_ix_{ij} - \cfrac{e^{X_i\theta}}{1+e^{X_i \theta}}x_{ij})$
## ${\partial\over{\partial \theta_j}}l(p)=-\Sigma((y_i - p_i)x_{ij})$

```python
def get_gradient_ij(X, y, parameters, j, model):
    if model == 'linear':
        y_hat = dot_product(X, parameters)
        gradient = -np.sum((y-y_hat)*X[j])
    else:
        p = p = logistic(X, parameters)
        gradient = -np.sum((y-p)*X[j])
    return gradient
```

## Batch Gradient
하나의 배치 (X_set, y_set)에 대해 기울기를 구하는 코드를 작성해주세요

```python
def batch_gradient(X_set, y_set, parameters, model):
    gradients = [0 for _ in range(len(parameters))]
    
    for i in range(len(X_set)):
        X = X_set.iloc[i,:]
        y = y_set.iloc[i]
        for j in range(len(parameters)):
            gradients[j] += get_gradient_ij(X, y, parameters, j, model)
    
    return np.array(gradients)/len(X_set)
```

## mini-batch
인덱스로 미니 배치 나누기

```python
def batch_idx(X_train, batch_size):
    N = len(X_train)
    nb = (N // batch_size)+1 #number of batch
    idx = np.array([i for i in range(N)])
    idx_list = [idx[i*batch_size:(i+1)*batch_size] for i in range(nb) if len(idx[i*batch_size:(i+1)*batch_size]) != 0]
    return idx_list
```

batch_idx 함수에 대한 설명을 batch_size와 함께 간략하게 작성해주세요  
### 설명
-  batch_idx는 총 개수를 batch_size로 나눠서 소분 한 뒤에, index를 붙여서 저장해둔 것이다. 이를 통해 소분한 것들중 몇 번째를 돌고 있는지 확인할 수 있다. 이 때 batch_size로 총 개수를 나눈 값이 0이 아닐 경우 남는 것들을 넣을 하나를 추가로 더해준다. 

## Update Parameters
기울기를 갱신하는 코드를 작성해주세요  
(loss와 마찬가지로 기울기를 갱신할 때 배치 사이즈를 고려해 평균으로 갱신해주세요)

```python
def step(parameters, gradients, learning_rate, n): #n:현재 배치의 데이터 수
    for i in range(len(parameters)):
        #경사를 업데이트 한다.
        gradients[i] *= learning_rate/n
    parameters -= gradients
    return parameters
```

## Gradient Descent
위에서 작성한 함수들을 조합해서 경사하강법 함수를 완성해주세요

- learning_rate: 학습률  
- tolerance: Step이 너무 작아서 더 이상의 학습이 무의미할 때 학습을 멈추는 조건  
- batch: 기울기를 1번 갱신할 때 사용하는 데이터셋  
- epoch:  
- num_epoch:
<br>


##### BGD
- Batch gradient descent
- batch_loss(X_batch, y_batch, step(parameters, batch_gradient(X_batch, y_batch, parameters, model), learning_rate, len(X_batch), lossfunction, len(X_batch))

##### SGD
- Stochastic gradient descent
- batch_loss(X_batch, y_batch, step(parameters, batch_gradient(X_batch, y_batch, parameters, model), learning_rate, 1), lossfunction, 1) 

##### MGD
- Minibatch gradient descent
- batch_loss(X_batch, y_batch, step(parameters, batch_gradient(X_batch, y_batch, parameters, model), learning_rate, batch_size), lossfunction, batch_size)  
<br>
batch_size에 따른 경사하강법의 종류를 적어주세요  
batch_size=1 ->  SGD
batch_size=k -> MGD
batch_size=whole -> BGD

```python
def gradient_descent(X_train, y_train, learning_rate = 0.1, num_epoch = 1000, tolerance = 0.00001, model = 'logistic', batch_size = 16):
    stopper = False
    
    N = len(X_train.iloc[0])
    parameters = np.random.rand(N)
    loss_function = minus_log_cross_entropy_i if model == 'logistic' else mse_i
    loss = 999
    batch_idx_list = batch_idx(X_train, batch_size)
    
    for epoch in range(num_epoch):
        if stopper:
            break
        for idx in batch_idx_list:
            X_batch = X_train.iloc[idx,]
            y_batch = y_train.iloc[idx]
            gradients = batch_gradient(X_batch, y_batch, parameters, model)
            parameters = step(parameters, gradients, learning_rate, len(idx))
            new_loss = batch_loss(X_batch, y_batch, parameters, loss_function, len(idx))
            
            #중단 조건
            if abs(new_loss - loss) < tolerance:
                stopper = True
                break
            loss = new_loss
        
        #100epoch마다 학습 상태 출력
        if epoch%100 == 0: #출력이 길게 나오면 check point를 수정해도 됩니다.
            print(f"epoch: {epoch}  loss: {new_loss}  params: {parameters}  gradients: {gradients}")
    
    return parameters
```

## Implement
### Hyperparameter를 수업 때 배웠던 방법 중 하나를 골라 최적화 시켜주세요
경사하강법 함수를 이용해 최적의 모수 찾아보세요. 학습을 진행할 때, Hyper Parameter를 바꿔가면서 학습시켜보세요.

```python
# 베이지안 최적화 (Bayesian Optimization)를 사용하였습니다.

# Bayesian Optimization 은 어느 입력값(x)를 받는 미지의 목적 함수(f(x))를 상정하여,
# 해당 함숫값(f(x))을 최대로 만드는 최적해를 찾는 것을 목적으로 합니다.
# 즉, 목적 함수(탐색대상함수)와 하이퍼파라미터 쌍(pair)을 대상으로 Surrogate Model(대체 모델) 을 만들고,
# 순차적으로 하이퍼 파라미터를 업데이트해 가면서 평가를 통해 최적의 하이퍼파라미터 조합을 탐색합니다.
# 이 때의 목점 함수를 black-box function 이라고 합니다.
# Bayesian Optimization 에는 두 가지 필수 요소가 존재합니다.

# 먼저 Surrogate Model 은, 현재까지 조사된 입력값-함숫결과값 점들 (x1, f(x1)),...,(xt, f(xt)) 을 바탕으로, 미지의 목적 함수의 형태에 대한 확률적인 추정을 수행하는 모델을 지칭합니다.
# 그리고 Acquisition Function 은, 목적 함수에 대한 현재까지의 확률적 추정 결과를 바탕으로, ‘최적 입력값을 찾는 데 있어 가장 유용할 만한’ 다음 입력값 후보를 추천해 주는 함수를 지칭합니다.
```

```python
def gradient_descent_loss(X_train, y_train, learning_rate = 0.1, num_epoch = 1000, tolerance = 0.00001, model = 'logistic', batch_size = 16):
    stopper = False
    
    N = len(X_train.iloc[0])
    parameters = np.random.rand(N)
    loss_function = minus_log_cross_entropy_i if model == 'logistic' else mse_i
    loss = 999
    batch_idx_list = batch_idx(X_train, batch_size)
    
    for epoch in range(num_epoch):
        if stopper:
            break
        for idx in batch_idx_list:
            X_batch = X_train.iloc[idx,]
            y_batch = y_train.iloc[idx]
            gradients = batch_gradient(X_batch, y_batch, parameters, model)
            parameters = step(parameters, gradients, learning_rate, len(idx))
            new_loss = batch_loss(X_batch, y_batch, parameters, loss_function, len(idx))
            
            #중단 조건
            if abs(new_loss - loss) < tolerance:
                stopper = True
                break
            loss = float(new_loss)
        
        #100epoch마다 학습 상태 출력
        if epoch%100 == 0: #출력이 길게 나오면 check point를 수정해도 됩니다.
            print(f"epoch: {epoch}  loss: {new_loss}  params: {parameters}  gradients: {gradients}")
    print(loss)
    return loss
```

```python
# !pip3 install bayesian-optimization
def get_loss(batch_size):
    loss = gradient_descent_loss(X_train, y_train,batch_size)
    return loss
```

```python
  #  bayesian-optimization 라이브러리의 BayesianOptimization 클래스 import
from bayes_opt import BayesianOptimization


# 실험해보고자하는 hyperparameter 집합
pbounds = {'batch_size': (10, 50)}

# Bayesian optimization 객체 생성
# f : 탐색 대상 함수, pbounds : hyperparameter 집합
# verbose = 2 항상 출력, verbose = 1 최댓값일 때 출력, verbose = 0 출력 안함
# random_state : Bayesian Optimization 상의 랜덤성이 존재하는 부분을 통제 
bo=BayesianOptimization(get_loss, pbounds=pbounds, verbose=2, random_state=1 )    

# 메소드를 이용해 최대화 과정 수행
# init_points :  초기 Random Search 갯수
# n_iter : 반복 횟수 (몇개의 입력값-함숫값 점들을 확인할지! 많을 수록 정확한 값을 얻을 수 있다.)
# acq : Acquisition Function들 중 Expected Improvement(EI) 를 사용
# xi : exploration 강도 (기본값은 0.0)
```

```python
# 메소드를 이용해 최대화 과정 수행
# init_points :  초기 Random Search 갯수
# n_iter : 반복 횟수 (몇개의 입력값-함숫값 점들을 확인할지! 많을 수록 정확한 값을 얻을 수 있다.)
# acq : Acquisition Function들 중 Expected Improvement(EI) 를 사용
# xi : exploration 강도 (기본값은 0.0)
bo.maximize(init_points=2, n_iter=10)

# ‘iter’는 반복 회차, ‘target’은 목적 함수의 값, 나머지는 입력값을 나타냅니다. 
# 현재 회차 이전까지 조사된 함숫값들과 비교하여, 현재 회차에 최댓값이 얻어진 경우, 
# bayesian-optimization 라이브러리는 이를 자동으로 다른 색 글자로 표시하는 것을 확인할 수 있습니다

# 찾은 파라미터 값 확인
print(bo.max)
```

```python
tmp = np.array([1 for _ in range(150)])
X = np.vstack((tmp, raw_X)).T
X = pd.DataFrame(X)
y = pd.Series(y)
```

### Estimation

```python
#정규방정식
theta = np.linalg.inv(np.dot(X.T,X)).dot(X.T).dot(y)
theta
```

```python
#경사하강법
new_param = gradient_descent(X, y, learning_rate = 0.1, num_epoch = 1000, tolerance = 0.00001, model = 'logistic', batch_size = len(X))
new_param
```

```python
y_hat_NE = theta.dot(X.T)
y_hat_GD = new_param.dot(X.T)
```
